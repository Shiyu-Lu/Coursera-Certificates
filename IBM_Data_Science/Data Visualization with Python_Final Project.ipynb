```python
import numpy as np
import pandas as pd

%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# read the data into a dataframe
survey = pd.read_csv('https://cocl.us/datascience_survey_data', index_col = 0)
survey.head()
```

```python
# Sort the dataframe in descending order of Very interested
survey.sort_values(by='Very interested', ascending=False, inplace=True)

# Convert the numbers into percentages of the total number of respondents. Recall that 2,233 respondents completed the survey. Round percentages to 2 decimal places.
for column in survey.columns:
    survey[column] = round(survey[column]/2233, 2)
survey.head()
```

```python
# Use a figure size of (20, 8)
# Bar width of 0.8
# Use color #5cb85c for the Very interested bars, color #5bc0de for the Somewhat interested bars, and color #d9534f for the Not interested bars
# Use font size 14 for the bar labels, percentages, and legend
# Use font size 16 for the title

survey.plot(kind='bar', figsize=(20, 8), width=0.8,
           color=['#5cb85c','#5bc0de','#d9534f'])

plt.xticks(rotation=-15)
plt.xlabel('Topics', fontsize=14)
plt.ylabel('Percentage', fontsize=14)
plt.legend(fontsize=14)
plt.title('Bar chart of the survey', fontsize=16)

# Display the percentages above the bars as shown above
xList = np.arange(len(survey.index)*3)
yList = []
for i in range(len(survey.index)):
    for j in range(3):
        yList.append(survey.iloc[i,j])
print(xList,yList)
for a, b in zip(xList, yList):
    plt.text(a/3.5-0.3,b+0.05,b,ha = 'center',va = 'bottom',fontsize=14)
    


# Remove the left, top, and right borders
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
```
